```python
import folium
from folium import plugins
```

```python
map_center = [30, 0] 
f1_map = folium.Map(location=map_center, zoom_start=2)
```

```python
for index, row in circuits.iterrows():
    folium.Marker([row['lat'], row['lng']], popup=row['name'], icon=folium.Icon(color='red')).add_to(f1_map)
```

```python
minimap = plugins.MiniMap()
f1_map.add_child(minimap)
```
