```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/aamirpirani/Desktop/Data_Analysis/Molecule_Solubility/Data/delaney_solubility_with_descriptors.csv')
```

```python
df.head()
```

# Data Preparation

## Data Seperation as X and Y

```python
y = df['logS']
y
```

```python
x = df.drop('logS', axis=1)
x
```

## Data Splitting

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
x_train
```

```python
x_test
```

# Model 

## Linear Regression

### Training Model

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression ()
lr.fit(x_train, y_train)
```

### Applying Model 

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_test_pred
```

```python
y_lr_train_pred
```

### Evaluate Model Performance

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

## Random Forest 

### Training Model 

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

### Applying the Model 

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

### Model Evaluation 

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
rf_results
```

## Model Comparison 

```python
df_models = pd.concat([lr_results, rf_results], axis=0)
df_models
```

```python
df_models.reset_index(drop=True)
```

## Data Visualization 

```python
import matplotlib.pyplot as plt 
import numpy as np

y_train = np.random.rand(50)
y_lr_train_pred = 0.5 * y_train + 0.2 * np.random.randn(50)

plt.style.use('dark_background')

plt.figure(figsize=(8, 6))
plt.scatter(x=y_train, y=y_lr_train_pred, c="#8A2BE2", alpha=0.3)

z = np.polyfit(y_train, y_lr_train_pred, 1)
p = np.poly1d(z)
plt.plot(y_train, p(y_train), 'red')

plt.ylabel('Predict LogS', color='white')
plt.xlabel('Experimental LogS', color='white')
plt.tick_params(axis='both', colors='white')

plt.show()
```
